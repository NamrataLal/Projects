# Simple Linear Regression Model
   Simple Linear Regression Model with independent variable movie budget and        dependent variable movie revenue.

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
np.set_printoptions(precision = 2, suppress = True)
```

```python
df = pd.read_csv('bud_vs_rev.csv', parse_dates = ['release_date'], index_col = 'release_date')
df = df.loc['2016']
df
```

```python
sns.set(font_scale = 1.5)
sns.lmplot(data = df, x = 'budget', y = 'revenue', ci = None)
plt.show()
```

```python
x = df.iloc[:,-2].values                    #linear regression model using numpy
x
```

```python
y = df.iloc[:,-1].values
y
```

```python
reg = np.polyfit(x = x, y = y, deg = 1)
reg
```

```python
np.polyfit(x = df.budget, y = df.revenue, deg = 1)
```

```python
X = np.array([min(x), max(x)])    #creating x and y values for regression line
X
```

```python
Y = np.polyval(reg, X)
Y
```

```python
plt.figure(figsize = (12,8))
plt.plot(X,Y)
plt.scatter(x= x,y = y)
plt.xlabel('Budget(in MUSD)')
plt.ylabel('Revenue(in MUSD)')
plt.show()
```
